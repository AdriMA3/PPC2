```python
import pandas as pd
link='https://github.com/AdriMA3/PC2/raw/refs/heads/main/HDI.xlsx'
datos=pd.read_excel(link)

len(datos)
```

```python
datos.head()
```

```python
#vemos que nuestra tabla esta para limpiar
#primero eliminaremos filas que no nos sirven
datos=datos.iloc[7:]
```

```python
datos
```

```python
#eliminaremos las filas donde en la columna Back esta vacia
datos=datos.dropna(subset=['Back'])
```

```python
datos
```

```python
#rename
datos.columns=['HDI_Rank','Country','Human_Development_Index','colum1','Life_expectancy_at_birth','colum2','Expected_years_of_schooling','colum3','Mean_years_of_schooling','colum4','Gross_national_income_per_capita','colum5','colum6','colum7','colum8']
```

```python
datos
```

```python
#elimar columnas que no nos sirven
datos=datos.drop(['HDI_Rank','colum1','colum2','colum3','colum4','colum5','colum6','colum7','colum8'],axis=1)
datos.head()
```

```python
#queremos que empiece de 1 no de 7
datos=datos.reset_index(drop=True)
datos.index=datos.index+1
datos.head()
```

```python
#veremos si esta en object las columnas
datos.dtypes
```

```python
columnas_redondeo=['Life_expectancy_at_birth','Expected_years_of_schooling','Mean_years_of_schooling','Gross_national_income_per_capita']
for col in columnas_redondeo:
  datos[col]=pd.to_numeric(datos[col],errors='coerce')
```

```python
#redondear algunas columnas
datos=datos.round({'Life_expectancy_at_birth':1,'Expected_years_of_schooling':1,'Mean_years_of_schooling':1,'Gross_national_income_per_capita':2})
datos.head()
```

```python
#pasamos Human_Development_Index de object a float64
datos['Human_Development_Index']=pd.to_numeric(datos["Human_Development_Index"],errors='coerce')
```

```python
datos.dtypes
```

```python
#creamos una columna y agrupamos como visto en clase
bins=[0,0.549,0.699,0.799,1.0]
labels=["LOW HUMAN DEVELOPMENT","MEDIUM HUMAN DEVELOPMENT","HIGH HUMAN DEVELOPMENT","VERY HIGH HUMAN DEVELOPMENT"]

datos["Category"]=pd.cut(datos["Human_Development_Index"],bins=bins,labels=labels)
datos.head()
```
